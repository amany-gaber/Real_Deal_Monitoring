```python
CATEGORIES = {
    "Electronics": {
        "MOBILES & ACCESSORIES": [
            "iPhones", "samsung" , 'relmi', "redmi" , "Budget Smartphones", "Tablets",
            "Headsets & Speakers", "Wearables", "Power Banks", "Chargers"
        ],
        "LAPTOPS & ACCESSORIES": [
            "MacBooks", "Gaming Laptops", "Monitors", "Printers",
            "Storage Devices", "Input Devices"
        ],
        "GAMING ESSENTIALS": [
            "Gaming Consoles", "Gaming Accessories", "Video Games",
            "Gaming Monitors", "Digital Cards"
        ],
        "TVS & HOME ENTERTAINMENT": [
            "LED", "QLED", "OLED", "4K", "8K",
            "Projectors", "Soundbars", "Streaming Devices"
        ],
        "CAMERAS": [
            "Action Cameras", "DSLR Cameras", "Surveillance Cameras",
            "Instant Cameras", "Camera Accessories"
        ],
    },

    "Women's Fashion": {
        "CLOTHING": ["Tops", "Dresses", "Pants", "Head Scarves", "Jeans", "Bodysuits"],
        "SPORTSWEAR": ["Tops", "Leggings", "Shorts", "Sport Bras", "Sport Shoes", "Sneakers"],
        "FOOTWEAR": ["Sports Shoes", "Sneakers", "Sandals", "Heels", "Flats", "Boots", "Flip Flops"],
        "BAGS & ACCESSORIES": ["Totes", "Shoulder Bags", "Cross-body Bags", "Wallets", "Jewelry", "Eyewear", "Watches"],
    },

    "Men's Fashion": {
        "CLOTHING": ["Shirts", "Polos", "Pants", "Jeans", "Sportswear"],
        "SPORTSWEAR": ["Tops", "Jackets", "Bottoms", "Sport Shoes", "Sneakers", "Track Pants"],
        "FOOTWEAR": ["Sports Shoes", "Sneakers", "Football Shoes", "Boots", "Flip Flops"],
        "BAGS": ["Backpacks", "Wallets", "Luggage", "Jewelry", "Belts", "Watches", "Eyewear"],
    },

    "Kids' Fashion": {
        "GIRLS CLOTHING": ["Tops", "Pants", "Clothing Sets", "Dresses", "Sportswear", "Jackets & Outerwear"],
        "BOYS CLOTHING": ["Tops", "Pants", "Clothing Sets", "Sweaters", "Sportswear", "Jackets & Outerwear"],
        "KIDS FASHION": ["Footwear", "Sports Shoes", "Sneakers", "Accessories", "Backpacks"],
    },

    "Beauty & Fragrance": {
        "MAKEUP": ["Mascaras", "Foundations", "Blushers", "Eyeshadows", "Lip Glosses", "Makeup Brushes", "Makeup Removers", "Concealers"],
        "SKINCARE": ["Moisturizers", "Suncare", "Bath & Body", "Cleansers", "Toners", "Serums", "Eye Creams"],
        "HAIRCARE": ["Shampoos", "Conditioners", "Hair Masks", "Hair Oils", "Hair Color", "Accessories"],
        "PERSONAL CARE": ["Bath & Body", "Oral Care", "Roll-ons & Deos", "Feminine Care", "Face Wash", "Lip Care"],
        "FRAGRANCE": ["Women’s Fragrances", "Men’s Fragrances", "Body Mists & Sprays"],
        "MEN’S GROOMING": ["Razors & Blades", "Shaving Creams & Gels", "Face Wash & Aftershaves", "Beard Care"],
        "HAIR STYLING TOOLS": ["Hair Dryers", "Straighteners", "Curling Irons", "Rollers", "Brushes"],
    },

    "Home & Appliances": {
        "Kitchen & Dining": ["Cookware", "Storage & Organisation", "Dinnerware & Serveware", "Kitchen Accessories", "Bakeware"],
        "Furniture": ["Coffee Tables", "Gaming Chairs", "Bean Bags", "Office Furniture", "Sofas", "Mattresses"],
        "Tools & Home Improvement": ["Power Tools", "Hand Tools", "Cleaning Supplies", "Home Organisation"],
        "Home Decor": ["Lighting", "Mats & Carpets", "Vases", "Mirrors"],
        "Bath & Bedding": ["Towels", "Bedding Sets", "Blankets"],
        "Small Appliances": ["Air Fryers", "Coffee Makers", "Blenders", "Vacuums"],
        "Large Appliances": ["Refrigerators", "Washing Machines", "Dishwashers", "Air Conditioners"],
    },

    "Baby": {
        "Baby Essentials": ["Diaper needs", "Skin & Bath care", "Feeding", "Car Seats & Strollers", "Baby clothing", "Safety equipment"],
        "Feeding Essentials": ["Breast Pumps", "Bottles", "Pacifiers", "Highchairs", "Baby Food"],
        "Baby Care": ["Diapers", "Wipes", "Grooming & Health"],
        "Nursery Essentials": ["Baby Furniture & Bedding"],
        "Baby Travel Gear": ["Strollers", "Car Seats", "Diaper Bags"],
    },

    "Toys & Games": {
        "Toys & Games": ["Toys for Girls", "Toys for Boys", "Party Supplies", "Puzzles", "Board Games", "Educational", "Outdoor Play", "Scooters", "Remote Control"]
    },

    "Supermarket": {
        "Supermarket": ["Home Care & Cleaning", "Beverages", "Canned & Packaged Food", "Herbs & Spices", "Snack Food", "Breakfast Food"]
    },

    "Automotive": {
        "Automotive": ["Oils & Fluids", "Interior Accessories", "Exterior Accessories", "Car Care", "Car Electronics"]
    },

    "Health & Nutrition": {
        "Health & Nutrition": ["Vitamins & Supplements", "Sexual Wellness", "Health Monitors", "Massage & Relaxation"]
    },

    "Sports & Outdoors": {
        "Sports & Outdoors": ["Exercise & Fitness", "Team Sports", "Racket Sports", "Cycling", "Skates & Scooters"]
    },

    "Stationery & Books": {
        "Stationery & Books": ["Paper", "Education & Crafts", "Desk Accessories", "Writing & Correction Supplies", "Office Electronics", "Stationery Essentials"]
    }
}


# ----------------- أدوات مساعدة -----------------
def safe_text(el):
    return el.get_text(strip=True) if el else None


def slugify(s: str) -> str:
    s = s.strip().lower()
    s = re.sub(r"[^\w\s-]", "", s)
    s = re.sub(r"[\s/]+", "_", s)
    s = re.sub(r"_+", "_", s)
    return s[:120]


def ensure_dir(path):
    os.makedirs(path, exist_ok=True)


def load_progress():
    if os.path.exists(PROGRESS_FILE):
        with open(PROGRESS_FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}


def save_progress(progress):
    with open(PROGRESS_FILE, "w", encoding="utf-8") as f:
        json.dump(progress, f, ensure_ascii=False, indent=2)




def to_number(x):
    if not x:
        return None
    x = x.replace(",", "").strip()
    try:
        return float(x)
    except:
        return None


def parse_products(html, page_no):
    soup = BeautifulSoup(html, "html.parser")
    rows = []
    titles = soup.select('h2[data-qa="plp-product-box-name"]')

    for h2 in titles:
        title = h2.get("title") or safe_text(h2)

        card = h2
        for _ in range(20):
            if card and card.name == "div" and (
                card.select_one('[data-qa="plp-product-box-price"]')
                or card.select_one('img[src*="nooncdn.com/p/"]')
            ):
                break
            card = card.parent

        if not card:
            continue

        a = h2.find_parent("a")
        if not a:
            a = card.select_one("a[href]")
        product_url = urljoin(BASE, a["href"]) if a and a.get("href") else None

        price_box = card.select_one('[data-qa="plp-product-box-price"]')
        currency = safe_text(price_box.select_one("span")) if price_box else None
        price = safe_text(price_box.select_one("strong")) if price_box else None
        old_price = safe_text(price_box.select_one('[class*="oldPrice"]')) if price_box else None

        txt = card.get_text(" ", strip=True)

        rating = None
        m = re.search(r"\b(\d\.\d)\b", txt)
        if m:
            rating = m.group(1)

        reviews = None
        count_span = card.select_one('[class*="countCtr"] span')
        if count_span:
            reviews = safe_text(count_span)
        else:
            m2 = re.search(r"\b(\d+(?:\.\d+)?[KkMm])\b", txt)
            if m2:
                reviews = m2.group(1)

        
        free_delivery = "Free Delivery" in txt
        noon_express = card.select_one('[data-qa="product-noon-express"]') is not None

        
        price_num = to_number(price)
        old_price_num = to_number(old_price)

       

        rows.append({
            "data": datetime.now(),
            "title": title,
            "product_url": product_url,
            "currency": currency,
            "price": price,
            "old_price": old_price,
            "rating": rating,
            "reviews": reviews,
            "free_delivery": free_delivery
        })

    return rows


def detect_total_pages_from_html(html):
    
    nums = []

    for m in re.finditer(r"[?&]page=(\d+)", html):
        try:
            nums.append(int(m.group(1)))
        except:
            pass

    if nums:
        return max(nums)

    return 1


def append_rows_to_csv(path, fieldnames, rows):
    file_exists = os.path.exists(path)
    with open(path, "a", newline="", encoding="utf-8") as f:
        w = csv.DictWriter(f, fieldnames=fieldnames)
        if not file_exists:
            w.writeheader()
        w.writerows(rows)


def open_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless=new")
    options.add_argument("--window-size=1400,900")
    options.add_argument("--lang=en-US")
    options.add_argument("--disable-blink-features=AutomationControlled")
    # يقلل مشاكل تحميل
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )
    return driver


def get_page_with_retry(driver, url, retries=5, base_wait=4):
    
    last_err = None
    for attempt in range(1, retries + 1):
        try:
            driver.get(url)
            time.sleep(base_wait)
            # scroll بسيط يساعد lazy load
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight * 0.6);")
            time.sleep(1.2)
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1.8)
            html = driver.page_source
            # لو الصفحة رجعت فاضية أو مفيش منتجات
            # لو الصفحة اتحملت حتى لو مفيهاش منتجات
            if "<html" in html.lower():
                return html

            last_err = "HTML not loaded properly"

        except Exception as e:
            last_err = str(e)

        time.sleep(2 * attempt)

    raise RuntimeError(f"Failed to load: {url}. Last error: {last_err}")


def scrape_query_to_file(driver, main_cat, sub_cat, item_query, progress):
    ensure_dir(OUT_DIR)

    query_key = f"{main_cat}||{sub_cat}||{item_query}"
    safe_name = slugify(f"{main_cat}_{sub_cat}_{item_query}")
    out_csv = os.path.join(OUT_DIR, f"{safe_name}.csv")

    fieldnames = [
        "date", "title", "product_url",
        "currency", "price", "old_price",
        "rating", "reviews",
        "free_delivery"
    ]

    # من أين نكمل؟
    start_page = progress.get(query_key, {}).get("last_page_done", 0) + 1

    # أول صفحة: نقرأ منها عدد الصفحات
    first_url = BASE_SEARCH.format(page=1, query=quote_plus(item_query))
    html1 = get_page_with_retry(driver, first_url)
    total_pages = detect_total_pages_from_html(html1)

    # لو كنا هنكمل من منتصف، نطبع العدد برضو
    print(f"\n[{main_cat} > {sub_cat} > {item_query}] total_pages={total_pages}, resume_from={start_page}")

    # لو start_page أكبر من total_pages خلاص
    if start_page > total_pages:
        print("  ✓ Already completed.")
        return

    # لو هنبدأ من 1: عالج الصفحة الأولى من html1 لتوفير وقت
    if start_page == 1:
        rows = parse_products(html1, 1)
        append_rows_to_csv(out_csv, fieldnames, rows)
        progress[query_key] = {"last_page_done": 1, "total_pages": total_pages}
        save_progress(progress)
        print(f"  page 1 -> {len(rows)} products saved")
        current = 2
    else:
        current = start_page

    # باقي الصفحات
    for p in range(current, total_pages + 1):
        url = BASE_SEARCH.format(page=p, query=quote_plus(item_query))
        html = get_page_with_retry(driver, url)
        rows = parse_products(html, p)

        # لو صفحة فاضية فجأة: نوقف (احتياط)
        if p != 1 and len(rows) == 0:
            print(f"  page {p} -> 0 products (stop)")
            progress[query_key] = {"last_page_done": p, "total_pages": total_pages}
            save_progress(progress)
            break

        append_rows_to_csv(out_csv, fieldnames, rows)

        progress[query_key] = {"last_page_done": p, "total_pages": total_pages}
        save_progress(progress)

        print(f"  page {p}/{total_pages} -> {len(rows)} products saved")

        # تهدئة لتقليل الحظر/الأعطال
        time.sleep(1.2)


def run_all(headless=False):
    progress = load_progress()
    driver = open_driver(headless=headless)

    try:
        for main_cat, subs in CATEGORIES.items():
            for sub_cat, items in subs.items():
                for item in items:
                    scrape_query_to_file(driver, main_cat, sub_cat, item, progress)
    finally:
        driver.quit()
        print("\nDONE ALL ✅")


if __name__ == "__main__":
    # خليها False أول مرة عشان تشوفي اللي بيحصل
    run_all(headless=False)
```
